# <span style="color:green; font-family:Georgia; font-size:1.5em;">4. Regression Lineaire Simple (Python) </span>


```python
# Todo 

# Estimation de \beta_1

beta1 = np.cov(overallQual, salePrice)[0,1]/np.var(overallQual)
print ("beta1 = ", beta1)
```

```python
# Todo 
# Estimation de \beta_0

# Calcul des moyennes de x et y
yMean = salePrice.mean()
xMean = overallQual.mean()
#print ("x,y = ", xMean, ",", yMean )

beta0 = yMean - beta1*xMean
print ("beta0 = ", beta0)

print ("==> y =", beta1, "x", beta0)

# fonction qui calcul y en fonction de x, beta0 et beta1
def f(x):
      return beta0 + beta1*x

plt.plot([f(x) for x in range(11)])

plt.show()
```

```python
# Calculez le coeffecient de détermination R2 
def SCE():
    somme = 0
    for i in overallQual:
        somme += (f(i) - yMean)**2
    return somme

def SCT():
    somme = 0
    for i in salePrice:
        somme += (i - yMean)**2
    return somme
    
R2 = SCE()/SCT()
print ("R2 = ", R2)
```

```python
# prédire les valeurs sur les données de Test
overallQualTest = houses_test.get('OverallQual')
salePriceTest = houses_test.get('SalePrice')

# Calcul du prix sur les données de test
pricePred = []
for i in overallQualTest:
    pricePred.append(f(i))

#pricePred[0]

```

# <span style="color:green; font-family:Georgia; font-size:1.5em;">5.  Evaluation du  modèle (RMSE)
  </span>


<span style="font-family:Georgia; font-size:1.2em;">La  racine  carrée  de l’erreur quadratique moyenne (RMSE) est une   autre  mesure  des  erreurs  de  prévision. La RMSE est définie par : $ RMSE = \sqrt{ \frac{1}{n} \sum (pred - y)^{2}} $.
</span>

<span style="font-family:Georgia; font-size:1.2em;">
Cette  mesure  pénalise  d'avantage  les  erreurs  importantes 
</span>

```python
# Calculez la RMSE de votre modèle sur les données de test

def CALCUL():
    somme = 0
    for i in range(0, len(pricePred)):
        somme += (pricePred[i] - salePriceTest[i])**2
    return somme


RMSE = np.sqrt(CALCUL()/len(pricePred))
print ("RMSE = ", RMSE)
```

# <span style="color:green; font-family:Georgia; font-size:1.5em;">5.  Pour aller plus loin </span>

<span style="font-family:Georgia; font-size:1.2em;">  
La standardisation des données consiste à centrer-réduire la matrice de fonctions caractéristiques  $X$ en retranchant pour chaque vecteur $X^{j}, j \in [1, N]$ sa moyenne et en divisant par son écart-type.
</span>


<span style="font-family:Georgia; font-size:1.2em;">
La formule d’usage est alors : $ ~~~~~~~~~~~  \forall j \in [1, N] ~~~~ X^{j} = \frac{X^{j}_{std} - \mu_{j}}{ \sigma^{2}_{j}}$

</span>

```python
# Todo :
##  refaire le calcul des paramètres après un prétraitement sur les données d'apprentissage

##  1.  Centrer la variable --> soustraire son espérance à chacune de ses valeurs initiales
overallQualCentrerReduite = []
for i in overallQual:
    overallQualCentrerReduite.append(i-xMean)  
    
##  2.  Réduire la variable --> diviser toutes ses valeurs par son écart type.
ecartType = np.std(overallQual)
for i in range(0, len(overallQualCentrerReduite)):
    overallQualCentrerReduite[i] = overallQualCentrerReduite[i]/ecartType
    
#overallQualCentrerReduite

## --------------------------
##    Soient  μ  l’espérance et σ l’écart-type des valeurs d’une variable aléatoire.
##    Centrer-réduire l’une de ses valeurs V revient alors à calculer : 
##          (V - μ ) / σ 
## --------------------------

##  3. re-éstimer les paramètre de modèle
# Estimation de \beta_1 et beta_0

beta1bis = np.cov(overallQualCentrerReduite, salePrice)[0,1]/np.var(overallQualCentrerReduite)
print ("beta1 = ", beta1bis)

xMeanbis = stat.mean(overallQualCentrerReduite)

beta0bis = yMean - beta1bis*xMeanbis
print ("beta0 = ", beta0bis)

print ("==> y =", beta1bis, "x", beta0bis)

# fonction qui calcul y en fonction de x, et des nouveaux parametres
def fbis(x):
      return beta0bis + beta1bis*x

plt.plot([fbis(x) for x in range(11)])
plt.show()


##  4. re-calculer la précision du modèle sur les données de test.
# Calculez le coeffecient de détermination R2 
def SCEbis():
    somme = 0
    for i in overallQualCentrerReduite:
        somme += (fbis(i) - yMean)**2
    return somme
    
R2 = SCEbis()/SCT()
print ("R2 = ", R2)

# Calcul du prix sur les données de test
pricePredbis = []
for i in overallQualTest:
    pricePredbis.append(fbis(i))

def CALCULbis():
    somme = 0
    for i in range(0, len(pricePredbis)):
        somme += (pricePredbis[i] - salePriceTest[i])**2
    return somme

RMSE = np.sqrt(CALCULbis()/len(pricePredbis))
print ("RMSE = ", RMSE)

```
